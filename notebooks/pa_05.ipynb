```python
# --- Import Libraries ---
import pandas as pd
import numpy as np

# TODO: Load your processed dataset (the CSV you saved at the end of the previous assignment)
df_clean = pd.read_csv('project_data_processed.csv')
df_clean.head()
df_clean.info()
df_clean.describe(include="all")
```

## Part 1: Grouping Data (1.5 Points)

Grouping allows you to segment your data and compute metrics (like mean, sum, or count) for each segment.

**Your Task:**
1.  Identify a categorical variable to group by.
2.  Apply an aggregation function to one or more numerical columns.
3.  If your dataset does not support grouping, explain the limitation in the markdown cell below.

```python
# TODO: Perform a groupby operation
# Example: df_grouped = df.groupby('category_column')['numeric_column'].mean()

# TODO: Display the result
# print(df_grouped.head())
category_summary = df_clean.groupby("product_category").agg({
    "total_revenue": "sum",
    "quantity_sold": "sum",
    "price": "mean",
    "rating": "mean"
}).reset_index()

category_summary
```

**Reflection:**
*Double click here to edit.*

I grouped the dataset by the categorical variable product_category and calculated several summary statistics for each category, including total revenue, total quantity sold, average price, and average rating. This aggregation summarizes overall performance within each product category rather than looking at individual transactions.

The grouped metrics provide insight into how different categories contribute to overall sales. For example, the results show that categories such as Beauty, Books, and Fashion generate similar levels of total revenue, each around 5.4–5.5 million. The quantity sold is also relatively similar across categories, suggesting that demand is fairly evenly distributed among product types in the dataset.

The average price and rating help provide additional context. Prices are fairly consistent across categories, averaging around $252, while ratings are close to 3 across all groups. This indicates that there are no large differences in pricing or customer satisfaction between product categories.

For the final analysis, this grouped information helps identify whether certain categories dominate revenue or sales and whether product characteristics differ across categories. It provides a higher level view of product performance that can guide further analysis, such as exploring how discounts, regions, or payment methods influence sales within specific categories.

## Part 2: Pivot Tables (1.5 Points)

Pivot tables provide a multi-dimensional summary of your data, allowing you to calculate metrics across two or more categorical axes.

**Your Task:**
1.  Construct a pivot table using `pd.pivot_table()`.
2.  Define the `index`, `columns`, and `values` parameters.
3.  If your dataset does not have suitable dimensions for a pivot table, explain the omission below.

```python
# TODO: Create a pivot table
# df_pivot = pd.pivot_table(df, values='numeric_val', index='category_1', columns='category_2', aggfunc='sum')

# TODO: Display the result
# print(df_pivot.head())
df_pivot = pd.pivot_table(
    df_clean,
    values="total_revenue",
    index="product_category",
    columns="customer_region",
    aggfunc="sum"
)

print(df_pivot)
```

**Reflection & Justification:**
*Double click here to edit.*

I created a pivot table using product_category as the index and customer_region as the columns, with total_revenue as the aggregated value. This structure allows the table to summarize how revenue is distributed across both product types and geographic regions simultaneously. Each row represents a product category, and each column represents a region. The values show the total revenue generated for that specific category–region combination.

This pivot table reveals how sales performance varies across markets. Overall, revenue is relatively evenly distributed across regions, with Asia, Europe, the Middle East, and North America each contributing similar amounts of revenue for most categories. However, small differences still appear. For example, Electronics generates slightly higher revenue in Europe, while Beauty and Home & Kitchen show strong revenue in North America. Books perform particularly well in the Middle East compared to other regions.

The table exposes the interaction between product demand and geographic markets, which would not be as clear using a single grouping variable. This type of summary is useful for identifying regional demand patterns and determining whether certain product categories perform better in specific areas. In the context of the final analysis, this helps highlight potential regional preferences and provides insight into where companies might focus marketing or inventory decisions.

**Your Task:**

```python
# TODO: Generate a cross-tabulation
# df_cross = pd.crosstab(df['category_A'], df['category_B'])

# TODO: Display the result
# print(df_cross.head())
df_cross = pd.crosstab(df_clean['product_category'], df_clean['payment_method_clean'])

print(df_cross)
```

**Reflection & Justification:**

The frequencies indicate that payment methods are fairly evenly distributed across categories, with no single payment type dominating purchases for any specific category. For example, Beauty, Books, Electronics, and other categories all show similar counts across payment methods such as Credit Card, Debit Card, UPI, Wallet, and Cash on Delivery. This suggests that customer payment preferences do not vary dramatically depending on the type of product purchased.

This cross tabulation helps provide a clearer understanding of purchasing behavior within the dataset. It shows that payment method usage is relatively balanced, indicating that customers across different product categories rely on a mix of payment options. For the final analysis, this suggests that payment method choice is likely not strongly tied to product category, but it still provides useful insight into the overall distribution of transaction types in the dataset.
